```python
import random

class Player:
    def __init__(self, name, batting_skill, bowling_skill, fielding_skill, running_skill, experience):
        self.name = name
        self.batting_skill = batting_skill
        self.bowling_skill = bowling_skill
        self.fielding_skill = fielding_skill
        self.running_skill = running_skill
        self.experience = experience

class Team:
    def __init__(self, name):
        self.name = name
        self.players = []
        self.runs = 0
        self.wickets = 0

    def add_player(self, player):
        self.players.append(player)

class Field:
    def __init__(self, size, fan_ratio, pitch_conditions, home_advantage):
        self.size = size
        self.fan_ratio = fan_ratio
        self.pitch_conditions = pitch_conditions
        self.home_advantage = home_advantage

class Umpire:
    def __init__(self):
        pass

    def predict_outcome(self, batsman, bowler):
        return random.choice([True, False])  # Randomly predict if the batsman is out

class Commentator:
    def __init__(self):
        pass

    def comment(self, batsman, bowler, runs_scored):
        print(f"Commentary: {batsman.name} is batting against {bowler.name}. {runs_scored} runs scored.")

class Match:
    def __init__(self, team1, team2, field):
        self.team1 = team1
        self.team2 = team2
        self.field = field
        self.umpire = Umpire()
        self.commentator = Commentator()

    def start_match(self):
        print("Match Started!\n")
        print(f"{self.team1.name} Batting:\n")
        self.play_innings(self.team1, self.team2)
        print("\nTeam Change: Innings Break\n")
        print(f"{self.team2.name} Batting:\n")
        self.play_innings(self.team2, self.team1)
        self.end_match()

    def play_innings(self, batting_team, bowling_team):
        for i in range(6):  # Play 6 overs
            print(f"Over {i+1}:")
            for j in range(6):  # Play 6 balls in each over
                batsman = random.choice(batting_team.players)
                bowler = random.choice(bowling_team.players)
                runs_scored = random.randint(0, 6)
                self.commentator.comment(batsman, bowler, runs_scored)
                batting_team.runs += runs_scored
                if self.umpire.predict_outcome(batsman, bowler):
                    batting_team.wickets += 1
                    print(f"{batsman.name} is out!")
                    if batting_team.wickets == 10:  # All out
                        print(f"All out! {batting_team.name} scored {batting_team.runs} runs for {batting_team.wickets} wickets.")
                        return
            print(f"{batting_team.name} scored {batting_team.runs} runs for {batting_team.wickets} wickets.")

    def end_match(self):
        print("Match Ended!")
        print(f"{self.team1.name} scored {self.team1.runs} runs for {self.team1.wickets} wickets.")
        print(f"{self.team2.name} scored {self.team2.runs} runs for {self.team2.wickets} wickets.")

        if self.team1.runs > self.team2.runs:
            print(f"{self.team1.name} wins!")
        elif self.team1.runs < self.team2.runs:
            print(f"{self.team2.name} wins!")
        else:
            print("It's a tie!")

# Create Players
player1 = Player("Rohit Sharma", 0.8, 0.2, 0.9, 0.8, 0.9)
player2 = Player("Shikhar Dhawan", 0.7, 0.1, 0.8, 0.9, 0.9)
player3 = Player("Virat Kohli", 0.9, 0.1, 0.9, 0.8, 0.9)
player4 = Player("KL Rahul", 0.8, 0.1, 0.7, 0.9, 0.8)
player5 = Player("Rishabh Pant", 0.7, 0.1, 0.7, 0.9, 0.8)
player6 = Player("Hardik Pandya", 0.8, 0.2, 0.8, 0.9, 0.8)
player7 = Player("Ravindra Jadeja", 0.6, 0.6, 0.9, 0.8, 0.9)
player8 = Player("Bhuvneshwar Kumar", 0.3, 0.8, 0.9, 0.7, 0.8)
player9 = Player("Jasprit Bumrah", 0.2, 0.8, 0.9, 0.5, 0.8)
player10 = Player("Yuzvendra Chahal", 0.1, 0.7, 0.8, 0.5, 0.8)
player11 = Player("Mohammed Shami", 0.1, 0.8, 0.9, 0.5, 0.8)

player12 = Player("David Warner", 0.8, 0.1, 0.8, 0.9, 0.8)
player13 = Player("Aaron Finch", 0.7, 0.1, 0.7, 0.9, 0.8)
player14 = Player("Steve Smith", 0.9, 0.1, 0.9, 0.8, 0.9)
player15 = Player("Glenn Maxwell", 0.8, 0.2, 0.9, 0.8, 0.8)
player16 = Player("Marcus Stoinis", 0.7, 0.3, 0.9, 0.7, 0.8)
player17 = Player("Alex Carey", 0.6, 0.1, 0.95, 0.7, 0.8)
player18 = Player("Mitchell Starc", 0.1, 0.8, 0.95, 0.5, 0.8)
player19 = Player("Adam Zampa", 0.1, 0.9, 0.97, 0.5, 0.8)
player20 = Player("Pat Cummins", 0.2, 0.8, 0.95, 0.5, 0.8)
player21 = Player("Josh Hazlewood", 0.1, 0.8, 0.9, 0.5, 0.8)
player22 = Player("Moises Henriques", 0.7, 0.3, 0.8, 0.7, 0.8)

# Create Teams
team1 = Team("India")
team2 = Team("Australia")

# Add Players to Teams
team1.add_player(player1)
team1.add_player(player2)
team1.add_player(player3)
team1.add_player(player4)
team1.add_player(player5)
team1.add_player(player6)
team1.add_player(player7)
team1.add_player(player8)
team1.add_player(player9)
team1.add_player(player10)
team1.add_player(player11)

team2.add_player(player12)
team2.add_player(player13)
team2.add_player(player14)
team2.add_player(player15)
team2.add_player(player16)
team2.add_player(player17)
team2.add_player(player18)
team2.add_player(player19)
team2.add_player(player20)
team2.add_player(player21)
team2.add_player(player22)

# Create Field
field = Field("Medium", 0.8, "Dry", "Australia")

# Create Match
match = Match(team1, team2, field)

# Start the match
match.start_match()
```
